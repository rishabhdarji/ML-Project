## Dataset Overview:

<br>

The Netflix Titles dataset, which is a complete collection of movies and series on Netflix and coverings different properties, such as type of titles, directors, cast, country of origin, year of release, rating, size, and the genres and a short description. With the help of this database, we are able to perform all the necessary to Netflix content coverage, genres popularity and distribution of shows within different time zones and the regions.


## Key Details:

<br>

Total Entries: The information comprises of 8809 entries, each containing an individual movie or a TV show.

Columns: The observations of the dataset are 12 columns as presented below:

show_id: A unique identifier with a number for every title.

type: It is the genre of the title as it can be either 'movie' or 'TV show'.

title: The first thing is name of the film or TV series.

director: The director (or directors) of the movie (or TV shows). (Does not typify all types of entries; particularly those of TV shows where similar rating might not be appropriate.)

cast: The actors/actresses list in the title underlines the main characters. (e.g. Some entries may miss out on this detail.)

country: The country or countries that the movie or show is produced in.

date_added: Release dates of the title by Netflix.

release_year: One of the most important facts about a motion picture is its release year.

rating: Classification into age groups of the title.

duration: The length of the title in length of minutes for movies and season if television shows.

listed_in: It is composes of the genres the title is a part of.

description: A brief summary of the title.

### Recommendation System

```python
from math import sqrt

def load_data(filepath):
    df = pd.read_csv(filepath)
    text_columns = ['description', 'listed_in', 'cast', 'director', 'release_year']
    df[text_columns] = df[text_columns].fillna('').astype(str)
    df['combined_features'] = df[text_columns].agg(' '.join, axis=1)
    return df

def create_term_frequency_matrix(df, column):
    vocabulary = {}
    tf_matrix = []
    
    for document in df[column]:
        tf_row = {}
        words = document.lower().split()
        for word in words:
            if word not in vocabulary:
                vocabulary[word] = len(vocabulary)
            if word in tf_row:
                tf_row[word] += 1
            else:
                tf_row[word] = 1
        tf_matrix.append(tf_row)
    
    return tf_matrix, vocabulary

def cosine_similarity(vector1, vector2):
    intersection = set(vector1.keys()) & set(vector2.keys())
    numerator = sum([vector1[x] * vector2[x] for x in intersection])
    
    sum1 = sum([vector1[x]**2 for x in vector1.keys()])
    sum2 = sum([vector2[x]**2 for x in vector2.keys()])
    denominator = sqrt(sum1) * sqrt(sum2)
    
    if not denominator:
        return 0.0
    else:
        return float(numerator) / denominator

def get_recommendations(tf_matrix, df, title, top_n=20):
    try:
        target_index = df[df['title'].str.lower() == title.lower()].index[0]
    except IndexError:
        return "Movie not found"
    
    scores = [cosine_similarity(tf_matrix[target_index], tf) for tf in tf_matrix]
    top_indices = sorted(range(len(scores)), key=lambda i: scores[i], reverse=True)[1:top_n+1]
    
    return df.iloc[top_indices]

def main():
    df = load_data("netflix_titles.csv")
    tf_matrix, vocabulary = create_term_frequency_matrix(df, 'combined_features')
    
    title = input("Enter a movie title: ")
    recommendations = get_recommendations(tf_matrix, df, title)
    
    if isinstance(recommendations, str):
        print(recommendations)
    else:
        print("Recommended shows similar to '{}':".format(title))
        title
        print(recommendations[['title', 'type', 'listed_in']])

if __name__ == "__main__":
    main()
```

```python

ground_truth = {
    'Zero': {'Jab Harry Met Sejal', 'Raees', 'Main Hoon Na'},
    'Big Miracle': {'Greater', 'Free Willy', 'The Lamb'},
    'A Mighty Team': {'Fatal', 'Forever Pure', 'The Last Whistle'}
}

recommendations = {
    'Zero': ['Main Hoon Na', 'Jab Harry Met Sejal', 'Raees'],
    'Big Miracle': ['Dr. Seuss The Lorax', 'Greater', 'The Lamb'],
    'A Mighty Team': ['Concrete Football', 'Forever Pure', 'The Last Whistle']
}

def calculate_evaluation_metrics(ground_truth, recommendations):
    tp = fp = fn = 0

    for user, correct_items in ground_truth.items():
        predicted_items = set(recommendations.get(user, []))
        
        tp_count = len(correct_items & predicted_items)
        fp_count = len(predicted_items - correct_items)
        fn_count = len(correct_items - predicted_items)
        
        tp += tp_count
        fp += fp_count
        fn += fn_count

    precision = tp / (tp + fp) if tp + fp > 0 else 0
    recall = tp / (tp + fn) if tp + fn > 0 else 0
    f_score = 2 * precision * recall / (precision + recall) if precision + recall > 0 else 0

    return precision, recall, f_score

precision, recall, f_score = calculate_evaluation_metrics(ground_truth, recommendations)
print(f"Precision: {precision * 100:.2f}")
print(f"Recall: {recall * 100:.2f}")
print(f"F-Score: {f_score * 100:.2f}")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def calculate_precision_recall_fscore(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    # True Positives 
    tp = np.sum((y_true == 1) & (y_pred == 1))
    # False Positives 
    fp = np.sum((y_true == 0) & (y_pred == 1))
    # False Negatives 
    fn = np.sum((y_true == 1) & (y_pred == 0))
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    return precision, recall, f1_score


def convert_duration(duration):
    if isinstance(duration, float):
        return 0  
    elif 'min' in duration:
        return int(duration.split()[0])
    elif 'Season' in duration:
        return int(duration.split()[0]) * 60  # Assuming each season is approximately 60 minutes
    else:
        return 0 


def manual_train_test_split(data, test_size=0.2, random_state=42):
    np.random.seed(random_state)
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_size)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]


def manual_standard_scaler(data):
    mean = data.mean()
    std = data.std()
    return (data - mean) / std


def manual_imputer(data):
    return data.fillna(data.mean())


def euclidean_distance(v1, v2):
    return np.sqrt(np.sum((v1 - v2) ** 2))


def get_neighbors(X_train, test_instance, k):
    distances = [euclidean_distance(x, test_instance) for x in X_train]
    indices = np.argsort(distances)[:k]
    return indices


def predict_classification(y_train, neighbor_indices):
    neighbor_labels = y_train.iloc[neighbor_indices]
    return neighbor_labels.mode()[0]

def knn_predict(X_train, y_train, X_test, k):
    predictions = []
    for test_instance in X_test:
        neighbor_indices = get_neighbors(X_train, test_instance, k)
        prediction = predict_classification(y_train, neighbor_indices)
        predictions.append(prediction)
    return predictions


netflix_data = pd.read_csv('netflix_titles.csv')  


titles = netflix_data.copy()
titles['num_cast'] = titles['cast'].str.split(',').apply(lambda x: len(x) if isinstance(x, list) else 0)
titles['num_genres'] = titles['listed_in'].str.split(',').apply(lambda x: len(x) if isinstance(x, list) else 0)
titles['duration_min'] = titles['duration'].apply(convert_duration)
titles['type'] = (titles['type'] == 'Movie').astype(int)
titles.drop(['show_id', 'title', 'director', 'cast', 'country', 'date_added', 'release_year', 'rating', 'duration', 'listed_in', 'description'], axis=1, inplace=True)


titles = manual_imputer(titles)


X = titles.drop('type', axis=1)
y = titles['type']
X_train, X_test = manual_train_test_split(X, test_size=0.2)
y_train, y_test = manual_train_test_split(y, test_size=0.2)
X_train_scaled = manual_standard_scaler(X_train)
X_test_scaled = manual_standard_scaler(X_test)
k = 10  # Number of neighbors
predictions = knn_predict(X_train_scaled.values, y_train, X_test_scaled.values, k)

def accuracy_metric(actual, predicted):
    correct = sum(1 for i in range(len(actual)) if actual[i] == predicted[i])
    return correct / float(len(actual)) * 100.0

accuracyknn = accuracy_metric(y_test.values, predictions)
print(f'accuracyknn: {round(accuracyknn, 2)}%')

precision, recall, f1_score = calculate_precision_recall_fscore(y_test.values, predictions)
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1-Score: {f1_score:.2f}')

def plot_confusion_matrix(y_true, y_pred):
    data = {'y_Actual':    y_true,
            'y_Predicted': y_pred
            }
    df = pd.DataFrame(data, columns=['y_Actual','y_Predicted'])
    confusion_matrix = pd.crosstab(df['y_Actual'], df['y_Predicted'], rownames=['Actual'], colnames=['Predicted'])
    sns.heatmap(confusion_matrix, annot=True, fmt='d')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()

plot_confusion_matrix(y_test.values, predictions)
```

### GaussianNaiveBayes

```python
import pandas as pd
import numpy as np

def calculate_precision_recall_fscore(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    # True Positives 
    tp = np.sum((y_true == 1) & (y_pred == 1))

    # False Positives 
    fp = np.sum((y_true == 0) & (y_pred == 1))

    # False Negatives 
    fn = np.sum((y_true == 1) & (y_pred == 0))

    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0

    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return precision, recall, f1_score

class GaussianNaiveBayes:
    def __init__(self):
        self.model = None

    def gaussian_probability(self, x, mean, std):
        exponent = np.exp(-((x - mean) ** 2 / (2 * std ** 2)))
        return (1 / (np.sqrt(2 * np.pi) * std)) * exponent

    def fit(self, X, y):
        labels = np.unique(y)
        data = {label: {} for label in labels}
        if not isinstance(X, pd.DataFrame):
            X = pd.DataFrame(X)
        for label in labels:
            features = X[y == label]
            data[label]["prior"] = len(features) / len(y)
            data[label]["summary"] = [(features[col].mean(), features[col].std()) for col in features.columns]
        self.model = data

    def predict(self, X):
        if not isinstance(X, pd.DataFrame):
            X = pd.DataFrame(X)
        results = []
        for _, instance in X.iterrows():
            probabilities = {}
            for label, parameters in self.model.items():
                total_probability = parameters["prior"]
                for idx, (mean, std) in enumerate(parameters["summary"]):
                    total_probability *= self.gaussian_probability(instance[idx], mean, std)
                probabilities[label] = total_probability
            results.append(max(probabilities, key=probabilities.get))
        return results

    def score(self, X, y):
        predictions = self.predict(X)
        return (predictions == y).mean() * 100

titles = netflix_data.copy()
titles['num_cast'] = titles['cast'].str.split(',').apply(lambda x: len(x) if isinstance(x, list) else 0)
titles['num_genres'] = titles['listed_in'].str.split(',').apply(lambda x: len(x) if isinstance(x, list) else 0)
titles['duration_min'] = titles['duration'].apply(convert_duration)
titles['type'] = (titles['type'] == 'Movie').astype(int)
titles.drop(['show_id', 'title', 'director', 'cast', 'country', 'date_added', 'release_year', 'rating', 'duration', 'listed_in', 'description'], axis=1, inplace=True)



X = titles.drop('type', axis=1)
y = titles['type']
gnb = GaussianNaiveBayes()
gnb.fit(X_train, y_train)
predictions = gnb.predict(X_test)
accuracygaussiannaivebyes = gnb.score(X_test, y_test)
print(f'accuracygaussiannaivebyes: {round(accuracygaussiannaivebyes, 2)}%')

def plot_confusion_matrix(y_true, y_pred):
    data = {'y_Actual':    y_true,
            'y_Predicted': y_pred
            }
    df = pd.DataFrame(data, columns=['y_Actual','y_Predicted'])
    confusion_matrix = pd.crosstab(df['y_Actual'], df['y_Predicted'], rownames=['Actual'], colnames=['Predicted'])
    sns.heatmap(confusion_matrix, annot=True, fmt='d')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()
    
precision, recall, f1_score = calculate_precision_recall_fscore(y_test.values, predictions)

print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1-Score: {f1_score:.2f}')

plot_confusion_matrix(y_test.values, predictions)
```

### XGB CLASSIFIER

```python
import numpy as np
import pandas as pd

def calculate_precision_recall_fscore(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    # True Positives
    tp = np.sum((y_true == 1) & (y_pred == 1))

    # False Positive
    fp = np.sum((y_true == 0) & (y_pred == 1))

    # False Negatives
    fn = np.sum((y_true == 1) & (y_pred == 0))
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return precision, recall, f1_score

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def gradient(y_true, y_pred):
    return y_true - sigmoid(y_pred)

def hessian(y_pred):
    p = sigmoid(y_pred)
    return p * (1 - p)

class DecisionTreeNode:
    def __init__(self):
        self.feature_index = None
        self.threshold = None
        self.left_value = None
        self.right_value = None

    def fit(self, X, gradients, hessians):
        min_loss = np.inf
        n_samples, n_features = X.shape
        
        for feature_index in range(n_features):
            for split_point in np.unique(X[:, feature_index]):
                left_mask = X[:, feature_index] < split_point
                right_mask = ~left_mask
                if not np.any(left_mask) or not np.any(right_mask):
                    continue
                
                left_score = np.sum(gradients[left_mask]) ** 2 / (np.sum(hessians[left_mask]) + 1e-10)
                right_score = np.sum(gradients[right_mask]) ** 2 / (np.sum(hessians[right_mask]) + 1e-10)
                total_gain = left_score + right_score
                
                if total_gain < min_loss:
                    min_loss = total_gain
                    self.feature_index = feature_index
                    self.threshold = split_point
                    self.left_value = np.sum(gradients[left_mask]) / (np.sum(hessians[left_mask]) + 1e-10)
                    self.right_value = np.sum(gradients[right_mask]) / (np.sum(hessians[right_mask]) + 1e-10)

    def predict(self, X):
        if isinstance(X, pd.DataFrame):
            X = X.values
        predictions = np.where(X[:, self.feature_index] < self.threshold, self.left_value, self.right_value)
        return predictions

class XGBoost:
    def __init__(self, n_estimators=100, learning_rate=0.1):
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.trees = []

    def fit(self, X, y):
        if isinstance(X, pd.DataFrame):
            X = X.values
        predictions = np.zeros(y.shape)
        for _ in range(self.n_estimators):
            residuals = gradient(y, predictions)
            h = hessian(predictions)
            tree = DecisionTreeNode()
            tree.fit(X, residuals, h)
            predictions += self.learning_rate * tree.predict(X)
            self.trees.append(tree)

    def predict_proba(self, X):
        if isinstance(X, pd.DataFrame):
            X = X.values
        tree_preds = [tree.predict(X) for tree in self.trees]
        final_preds = np.sum(tree_preds, axis=0)
        return sigmoid(final_preds * self.learning_rate)

    def predict(self, X):
        probabilities = self.predict_proba(X)
        return (probabilities >= 0.5).astype(int)


model = XGBoost(n_estimators=10, learning_rate=0.1)
model.fit(X_train, y_train)

predictions = model.predict(X_test)
accuracyxgbclassifier = np.mean(predictions == y_test)
print(f'accuracyxgbclassifier: {accuracyxgbclassifier * 100:.2f}%')

precision, recall, f1_score = calculate_precision_recall_fscore(y_test.values, predictions)

print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1-Score: {f1_score:.2f}')

def plot_confusion_matrix(y_true, y_pred):
    data = {'y_Actual':    y_true,
            'y_Predicted': y_pred
            }
    df = pd.DataFrame(data, columns=['y_Actual','y_Predicted'])
    confusion_matrix = pd.crosstab(df['y_Actual'], df['y_Predicted'], rownames=['Actual'], colnames=['Predicted'])
    sns.heatmap(confusion_matrix, annot=True, fmt='d')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()

plot_confusion_matrix(y_test.values, predictions)
```

```python
import xgboost as xgb
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import recall_score, accuracy_score, f1_score, confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

def classifier_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    
    accuracy = accuracy_score(y_test, y_pred) * 100
    recall = recall_score(y_test, y_pred, average='macro') * 100
    f1 = f1_score(y_test, y_pred) * 100
    conf_matrix = confusion_matrix(y_test, y_pred)
    
    
    return {
    'model': type(model).__name__,
    'accuracy': accuracy,
    'recall': recall,
    'f1_score': f1,
    'confusion_matrix': conf_matrix
}

models = [
    GaussianNB(),
    RandomForestClassifier(n_estimators=100, max_depth=10, random_state=0),
    KNeighborsClassifier(n_neighbors=5),
    xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss'),  # Updated XGBoost to suppress warnings
    tree.DecisionTreeClassifier()
]

results = []
for model in models:
    result = classifier_model(model, X_train, X_test, y_train, y_test)
    results.append(result)

results_df = pd.DataFrame(results)

print(results_df[['model', 'accuracy', 'recall', 'f1_score']])

for result in results:
    print(f"Confusion Matrix for {result['model']}:")
    sns.heatmap(result['confusion_matrix'], annot=True, fmt="d", cmap="Blues", xticklabels=['Movie', 'TV Show'], yticklabels=['Movie', 'TV Show'])
    plt.title('Confusion Matrix for ' + result['model'])
    plt.show()
```
